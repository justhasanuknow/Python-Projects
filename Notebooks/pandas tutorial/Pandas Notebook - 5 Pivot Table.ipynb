# Pivot Table

Pivot allows us to alter the dataframes

```python
import pandas as pd
```

```python
df = pd.read_excel("Excel and CSV Files/General Weather Data.xlsx")
df
```

```python
df.pivot(index = "date", columns = "city")
```

Pivot table is used to summarize and aggregate data inside dataframe.

```python
df.pivot_table(index = "city", columns = "date",aggfunc = "sum", values = "windspeed")
```

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html
<br>https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html


```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.pivot_table(index = pd.Grouper(freq = "M", key = "date"), columns = "city")
```
